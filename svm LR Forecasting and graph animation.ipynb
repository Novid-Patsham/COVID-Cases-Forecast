```python
import pandas as pd
from matplotlib  import pyplot as plt
import numpy as np
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
import random 
import math 
import time 

# for linear regression and support vector machine

from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.svm import SVR       
from sklearn.metrics import mean_squared_error, mean_absolute_error
import operator
warnings.filterwarnings("ignore", category=DeprecationWarning)
plt.style.use('seaborn')
```

```python
svmLR_data = pd.read_csv('Data\\india_brazil_cases.csv')

svmLR_data
```

```python
svmLR_data = svmLR_data[:132]
svmLR_data.Date = pd.date_range(start = '17/05/2021', periods = len(svmLR_data))
```

```python
# We will take the same daterange as Prophet (The last 2 months as training)

svmLR_IND = svmLR_data[['Date','Cases in India']][72:132]
svmLR_BRA = svmLR_data[['Date','Cases in Brazil']][72:132]

svmLR_BRA.reset_index(drop = True, inplace=True)
svmLR_IND.reset_index(drop = True, inplace=True)
```

```python
svmLR_BRA.head()
```

```python
# Indexing the days and generating future dates similar to a prophet object.

days_since_28_7_2021 = np.array([i for i in range(len(svmLR_BRA))]).reshape(-1, 1)
days_since_28_7_2021[-5:]
```

```python
# len(days_since_28_7_2021)
```

```python
B_cases = np.array(svmLR_BRA['Cases in Brazil']).reshape(-1, 1)
I_cases = np.array(svmLR_IND['Cases in India']).reshape(-1, 1)
```

## Brazil Preds

```python
# days in int format to fit in svm

days_in_future = 20
future_forcast = np.array([i for i in range(len(days_since_28_7_2021) + days_in_future)]).reshape(-1, 1)
adjusted_dates = future_forcast[:-15]
```

```python
# future_forcast[-5:]
```

```python
# days in datetime format for prediction

forecast_dates = pd.date_range(start = '07/28/2021', periods = 80)
forecast_dates[-5:]
```

```python
X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since_28_7_2021, B_cases, test_size=0.15, shuffle=False) 
```

#### SVM

```python
# building svm model

kernel = ['poly','sigmoid','rbf']
c = [0.01,0.1,1,10]
gamma = [0.01, 0.1, 1]
epsilon = [0.01, 0.1, 1]
shrinking = [True, False]
svm_grid = {'kernel':kernel, 'C':c , 'gamma': gamma, 'epsilon':epsilon, 'shrinking': shrinking }

svm = SVR()
svm_search = RandomizedSearchCV(svm, svm_grid, scoring = 'neg_mean_squared_error', cv = 3, return_train_score = True, n_jobs=-1, n_iter = 40,verbose =1)
svm_search.fit(X_train_confirmed, y_train_confirmed)
```

```python
svm_search.best_params_
```

```python
svm_confirmed = svm_search.best_estimator_
svm_pred = svm_confirmed.predict(future_forcast)
```

```python
# len(svm_pred)
```

```python
svm_test_pred = svm_confirmed.predict(X_test_confirmed)
plt.plot(svm_test_pred)
plt.plot(y_test_confirmed)
print('MAE:', mean_absolute_error(svm_test_pred, y_test_confirmed))
print('MSE:',mean_squared_error(svm_test_pred, y_test_confirmed))
```

#### Linear Regression

```python
linear_model = LinearRegression(normalize=True, fit_intercept=True)
linear_model.fit(X_train_confirmed, y_train_confirmed)
test_linear_pred = linear_model.predict(X_test_confirmed)
linear_pred = linear_model.predict(future_forcast)
print('MAE:', mean_absolute_error(test_linear_pred, y_test_confirmed))
print('MSE:',mean_squared_error(test_linear_pred, y_test_confirmed))
print('coef: ',linear_model.coef_)
print('intercept: ',linear_model.intercept_)
plt.plot(y_test_confirmed)
plt.plot(test_linear_pred)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import PyQt5
%matplotlib qt

x1 = forecast_dates[:-20]
x2 = forecast_dates
y = B_cases
z = svm_pred
x = linear_pred

fig, ax = plt.subplots(figsize=(20, 12))
line1, = ax.plot(x1, y, c='green', marker='o', alpha=.3, ms=15, lw=4)
line2, = ax.plot(x2, z, linestyle='dashed', color='blue')
line3, = ax.plot(x2, x, linestyle = 'dashed', color = 'black', lw = 2)

def update(num, x1,x2, y, z,x, line1, line2, line3):
    if num < len(x2):
        line1.set_data(x1[:num], y[:num])
        line2.set_data([], [])
        line3.set_data([], [])
    elif num < 2*len(x2):
        line2.set_data(x2[:num - len(x2)], z[:num - len(x2)])
        line3.set_data([], [])
    else:
        line3.set_data(x2[:num - 2*len(x2)], x[:num - 2*len(x2)])
    return [line1, line2 , line3]

ani = animation.FuncAnimation(fig, update, 3 * len(x2), fargs=[x1,x2, y, z,x, line1, line2, line3],
                              interval=20, repeat = False, blit=True)


plt.title('SVM and LR case prediction', size=30)
plt.xlabel('Days Since 08/07/2021', size=20)
plt.ylabel('# of Cases', size=30)
plt.legend(['Confirmed Cases', 'SVM predictions','LR predictions'], loc = 'upper left',frameon = True, fancybox =True, shadow = True, edgecolor = 'black',fontsize = 'xx-large', title = 'Brazil data',title_fontsize = 'xx-large')
plt.xticks(size=15)
plt.yticks(size=15)

plt.show()
```
